P(X = k) is the probability of selling k cars in a week,

P(sell some cars) = 1 - P(X = 0)

P(X = 0) = (e^(-3) * 3^0) / 0! = (e^(-3) * 1) / 1 ≈ 0.0498

P(sell some cars) = 1 - 0.0498 ≈ 0.9502

P(X = 3) = (e^(-3) * 3^3) / 3! ≈ 0.2240

Probability of selling 4 cars (P(X = 4)):

P(X = 4) = (e^(-3) * 3^4) / 4! ≈ 0.1680

```python
#(3C)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average number of cars sold per week
λ = 3

# Define the range of number of cars sold per week (0 to 10)
num_cars_sold = np.arange(0, 11)

# Calculate the cumulative probability using the Poisson CDF
cumulative_prob = poisson.cdf(num_cars_sold, λ)

# Plot the cumulative probability vs. number of cars sold per week
plt.plot(num_cars_sold, cumulative_prob, 'bo', label='Cumulative Probability')
plt.vlines(num_cars_sold, 0, cumulative_prob, colors='b', lw=2, alpha=0.5)
plt.xlabel('Number of Cars Sold per Week')
plt.ylabel('Cumulative Probability')
plt.title(' Cumulative Probability vs. Number of Cars Sold per Week(Applying Poissons Distribution Concept)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

```python
# B. Convert the ‘Funding’ features to a numerical value
def convert_funding_to_numeric(funding):
    if funding[-1] == 'K':
        return float(funding[1:-1]) / 1000
    elif funding[-1] == 'B':
        return float(funding[1:-1]) * 1000
    else:
        return float(funding[1:-1])

Company['Funds_in_million'] = Company['Funding'].apply(convert_funding_to_numeric)
```

```python
# A. Two-sample t-test
funds_operating = Company[Company['OperatingState'] == 'Operating']['Funds_in_million']
funds_closed_down = Company[Company['OperatingState'] == 'Closed']['Funds_in_million']

# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(funds_operating, funds_closed_down, equal_var=False)

# Print the t-statistic and p-value
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Compare the p-value to the significance level (e.g., 0.05)
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in the means of funds raised between companies that are still operating and companies that have closed down.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in the means of funds raised between companies that are still operating and companies that have closed down.")


```

```python
# B. Write the null hypothesis and alternative hypothesis.

# Null Hypothesis (H0): There is no significant difference in the means of funds raised between companies that are still operating and companies that have closed down.
# Alternative Hypothesis (Ha): There is a significant difference in the means of funds raised between companies that are still operating and companies that have closed down.
```

```python
# C. Test for significance and conclusion
# Perform the two-sample t-test
t_statistic, p_value = stats.ttest_ind(funds_operating, funds_closed_down, equal_var=False)

# Print the t-statistic and p-value
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Compare the p-value to the significance level (e.g., 0.05)
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. There is a significant difference in the means of funds raised between companies that are still operating and companies that have closed down.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference in the means of funds raised between companies that are still operating and companies that have closed down.")
```
